### Streaming & Batch

### Streaming:
Most models can stream thheir output content while it is being generated. By displaying output progressively, streaming significantly improves user experience, particularly for long responses.Calling stream() returns an iterator that yields output chunks as they are produced. You can use a loop to process each chunk in real time.

```python
model.stream("Write me a 200 words paragraph on Artificial Intelligence")
```

```python
for chunk in model.stream("Write me a 200 words paragraph on Artificial Intelligence"):
    print(chunk.text, end = "|", flush = True)
```

```python
for chunk in model.stream("Wy do parrots have colorful feathers?"):
    print(chunk.text, end = "|", flush = True)
```

### Batch:

Batching a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parallel.

```python
responses = model.batch([
    "Why do parrots have colorful feathers?",
    "Why do airplanes fly?",
    "What is Quantum Computing?"
])

for response in responses:
    print(response)
```

```python
model.batch([
    "Why do parrots have colorful feathers?",
    "Why do airplanes fly?",
    "What is Quantum Computing?",
    "What is Artificial Intelligence?",
    "Why do humans talk?",
    "Why humans can't communicate with animals?",
    "What is the formula of area?"],
    config = {
        'max_concurrency' : 5, # limit to 5 parallel calls
    }
)
```
